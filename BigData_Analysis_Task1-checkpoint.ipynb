```python
import dask.dataframe as dd

# Load your uploaded CSV
df = dd.read_csv("fake_sales_data.csv")

# Preview first few rows
df.head()
```

```python
# Check shape of the data
df.shape

# See column names
df.columns

# Summary statistics
df.describe().compute()
```

```python
# Check null values
df.isnull().sum().compute()
```

```python
# Descriptive statistics
df.describe().compute()
```

```python
# Create new column for total amount
df["total_amount"] = df["price"] * df["quantity"]
df[["customer_id", "total_amount"]].head()
```

```python
# Top 10 customers based on purchase amount
df_total = df[["customer_id", "total_amount"]].compute()
df_total = df_total.groupby("customer_id").sum().sort_values("total_amount", ascending=False)
df_total.head(10)
```

```python
# Average quantity purchased
df["quantity"].mean().compute()
```
